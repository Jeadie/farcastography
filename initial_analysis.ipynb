```python
import json 
import dataclasses

@dataclasses.dataclass
class Node:
    fid: int
    avatar: str
    fullname: str
    username: str
    idx: int # Set via enumeration of `nodes.json`.

with open("nodes.json") as f:
    nodes = dict([(n.get("fid"), Node(**n, idx=i)) for (i, n) in enumerate(json.load(f))])

with open("edges.json") as f:
    edges = json.load(f)
```

```python
import numpy as np

n = 500  # Number of Fid in nodes.json, not highest
adj = np.zeros((n, n))  # Default from-to self edge 

i = 0
for e in edges:
    adj[nodes[e["from"]].idx][nodes[e["to"]].idx]= 1.0
```

```python
d_in = np.sum(adj, axis=0)
d_out = np.sum(adj, axis=1)

L_in = d_in - adj
L_out = d_out - adj

D_in_inv_sqrt = np.sqrt(np.linalg.pinv(np.diag(d_in)))
D_out_inv_sqrt = np.sqrt(np.linalg.pinv(np.diag(d_out)))

# # Compute the normalized Laplacian: I - D^-0.5 * A * D^-0.5
L_norm = np.identity(adj.shape[0]) - np.dot(D_out_inv_sqrt, np.dot(adj, D_in_inv_sqrt))

values, vectors = np.linalg.eig(L_norm)
```

```python
import matplotlib.pyplot as plt
plt.hist(np.abs(values)[np.abs(values) < 5000], bins=25, edgecolor='black')
plt.title('Histogram of Eigenvalue Magnitudes < 5000')
plt.xlabel('Magnitude')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
k=100
embedding = vectors[:, np.argsort(np.abs(values))[:k]]

plt.hist(np.linalg.norm(embedding, axis=1), bins=25, edgecolor='black')
plt.title('Magnitude of Eigenvectors')
plt.xlabel('Magnitude')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
(np.linalg.norm(vectors, axis=1)).shape
```

```python
from sklearn.manifold import TSNE
embedding_2d = TSNE(n_components=2, learning_rate='auto', init='random', perplexity=3).fit_transform(np.abs(embedding))
embedding_2d.shape
```

```python
# Plot the points
plt.figure(figsize=(10, 6))
plt.scatter(embedding_2d[:,0], embedding_2d[:, 1], marker='o', edgecolor='k', s=50, alpha=0.6)
plt.xlabel('X Coordinate')
plt.ylabel('Y Coordinate')
plt.title('2D Coordinate Plot')
plt.grid(True)
plt.show()
```
